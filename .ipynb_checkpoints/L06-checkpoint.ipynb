### ЗАДАНИЕ 1

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
```

```python
boston = load_boston()

boston.keys()
```

```python
X = pd.DataFrame(boston['data'], columns=boston['feature_names'])
X.head(8)
```

```python
Y = pd.DataFrame(boston['target'], columns=["price"])
Y.head()
```

```python
X.info()
```

```python
Y.info()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, Y_train)
Y_lin_pred = lr.predict(X_test)
Y_lin_pred.shape
```

```python
from sklearn.metrics import r2_score

r2_lin = r2_score(Y_lin_pred.flatten(), Y_test)
r2_lin
```

### ЗАДАНИЕ 2

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
model.fit(X_train, Y_train.values[:, 0])
Y_forest_pred = model.predict(X_test)

r2_forest = r2_score(Y_forest_pred, Y_test)
r2_forest
```

##### Модель RandomForestRegressor работает лучше, т.к. R2 для неё ближе к единице, чем при использовании модели линейной регрессии (0.85 vs 0.67)

### ЗАДАНИЕ 4*

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# train_test_split уже ранее импортирована
```

```python
df = pd.read_csv('./creditcard.csv', sep=',')
pd.options.display.max_columns = 100
df.head(10)
```

```python
df.value_counts('Class', normalize=True)
```

```python
df.info()
```

```python
feature_names = df.columns.tolist()
feature_names.remove('Class')

XX = pd.DataFrame(df, columns=feature_names)
YY = df['Class']
XX.info()
YY.head(3)
```

```python
XX_train, XX_test, YY_train, YY_test = train_test_split(XX, YY, test_size=0.3, random_state=100, stratify=YY)
print(f"XX_train {XX_train.shape}\nYY_train {YY_train.shape}\nXX_test  {XX_test.shape}\nYY_test  {YY_test.shape}")
```

```python
parameters = {
    'n_estimators': [10, 15],
    'max_features': np.arange(3, 5),
    'max_depth': np.arange(4, 7),
}

clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3,
)

clf.fit(XX_train, YY_train)
```

```python
clf.best_params_
```

```python
YY_pred_test = clf.predict_proba(XX_test)
YY_pred_proba = YY_pred_test[:, 1]
YY_pred_proba
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(YY_test, YY_pred_proba)
```

```python
Xwine = pd.DataFrame(data['data'], columns=boston['feature_names'])
Xwine.info()
Xwine.shape
Xwine.head(5)
```

```python
Ywine = pd.DataFrame(data['target'])
Ywine.info()
Ywine.value_counts()
```

```python
Xwine['target'] = Ywine
Xwine.head(5)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
%config InlineBackend.figure_format = 'svg'

plt.figure(figsize = (12, 8))

sns.set(font_scale=0.8)

Xcorr = Xwine.corr()
Xcorr = np.round(Xcorr, 2)
Xcorr[np.abs(Xcorr) < 0.3] = 0

sns.heatmap(Xcorr, annot=True, linewidths=.5, cmap='YlOrBr')

plt.title('Correlation matrix')
plt.show()
```

```python
a = Xcorr['target']
high_corr = []
for i in range(len(a) - 1):
    if abs(a[i]) > 0.5:
        high_corr.append(Xcorr.columns[i])
    else:
        continue
high_corr
```

```python
#вариант расчёта high_corr
high_corr_rpt = list(Xcorr.loc[(abs(Xcorr['target']) > 0.5) & (abs(Xcorr['target']) < 1)].index)
high_corr_rpt
```

```python
Xwine2 = Xwine.copy()
Xwine2.drop(["target"], axis=1, inplace=True)
Xwine2
```

```python
new_colums_names = ['CHAS_2', 'RM_2', 'AGE_2', 'PTRATIO_2', 'B_2', 'LSTAT_2']
X_squares = Xwine2[high_corr]**2
X_squares.rename(columns = dict(zip(high_corr, new_colums_names)), inplace = True)
X_squares
```

```python
Xwine_upd = pd.concat([Xwine2, X_squares], axis=1)
Xwine_upd
```
