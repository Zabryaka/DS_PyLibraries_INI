```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline

%config InlineBackend.figure_format = 'svg'

plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('./creditcard.csv', sep=',')
df.head(5)
```

```python
df.shape
```

```python
cc_count = df['Class'].value_counts()
cc_count
```

```python
plt.bar(['0', '1'], cc_count)
plt.show()
```

```python
plt.bar(['0', '1'], cc_count)
plt.yscale(value="log")
plt.show()
```

```python
cc_v1 = df['V1']
cc_v1.index = df['Class']
cc_v1
```

```python
plt.hist(cc_v1[0], bins=20, density=True, alpha=0.5, color='Grey', ec='White')
plt.hist(cc_v1[1], bins=20, density=True, alpha=0.5, color='Red', ec='White')

plt.xlabel('V1')
plt.ylabel('Density')
plt.legend(labels=['Class 0', 'Class 1'], frameon=False)

plt.show()
```
